## Descripción del Problema

**Experimento base:** Simular el lanzamiento de dos dados justos de 6 caras.

**Preguntas:**

**a.** Estimar la probabilidad de que la suma sea igual a 7.

**b.** Estimar la probabilidad de que la suma sea 7 dado que al menos uno de los dados es par.

## Salidas Esperadas

- `P(suma = 7) = 0.1642`  
  (valor exacto: $\frac{1}{6} \approx 0.1666$)

- `P(suma = 7 | al menos un par) = 0.2174`  
  (valor exacto: $\frac{2}{9} \approx 0.2222$)

## Requisitos de Implementación

1. Fijar semilla aleatoria con valor **2026** antes de cualquier generación de números aleatorios.
2. Realizar **10,000 repeticiones** del experimento.
3. Imprimir respuestas como número decimal con **cuatro decimales**.

## Solución

### Importacion y semilla

```python
import random
import numpy as np

random.seed(2026)
N_SIMULACIONES = 10000
```

## Experimento 

Simular 2 dados justos de 6 caras, 10,000 repeticiones

```python
lista_dado1 = []
lista_dado2 = []

for _ in range(N_SIMULACIONES):
    lista_dado1.append(random.randint(1, 6))
    lista_dado2.append(random.randint(1, 6))

dado1 = np.array(lista_dado1)
dado2 = np.array(lista_dado2)
```

## INCISO A 

- `P(suma = 7) = 0.1642`  
  (valor exacto: $\frac{1}{6} \approx 0.1666$)

```python
incisoA = np.mean(dado1 + dado2 == 7)

print(f"P(suma = 7) = {incisoA:.4f}")
```

## INCISO B 

- `P(suma = 7 | al menos un par) = 0.2174`  
  (valor exacto: $\frac{2}{9} \approx 0.2222$)

```python
# evento A = suma = 7
# evento B = al menos un par

eventoB = (dado1 % 2 == 0) | (dado2 % 2 == 0)

probB = np.mean(eventoB)
probA = incisoA

# evento A = dado1 + dado2 == 7

prob_interseccion = np.mean((dado1 + dado2 == 7) & (eventoB))

incisoB = prob_interseccion / probB

print(f"P(suma =7 | al menos un par) = {incisoB:.4f}")
```

```python
import random
import numpy as np


N_SIMULACIONES = 10000
```

```python
random.seed(2026)

lista_dado1 = []
lista_dado2 = []
lista_dado3 = []

for _ in range(N_SIMULACIONES):
    lista_dado1.append(random.randint(1, 6))
    lista_dado2.append(random.randint(1, 6))
    lista_dado3.append(random.randint(1, 6))

dado1 = np.array(lista_dado1)
dado2 = np.array(lista_dado2)
dado3 = np.array(lista_dado3)


# EVENTO A: AL MENOS UN 6 

eventoA = (dado1 == 6) | (dado2 == 6) | (dado3 == 6)
probA = np.mean(eventoA)

# EVENTO B: la suma sea mayor a 12 
eventoB = (dado1 + dado2 + dado3) > 12
probB = np.mean(eventoB)

inciso1 = np.mean(eventoA & eventoB) / probB

print(f"P(al menos un 6 | suma > 12) = {inciso1:.4f}")
```
