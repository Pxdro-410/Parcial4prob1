## Descripción del Problema

**Experimento base:** Simular el lanzamiento de tres monedas justas (cara o cruz).

**Preguntas:**

**a.** Estimar la probabilidad de obtener exactamente dos caras.

**b.** Sea $X$ el número de caras. Estimar $E[X]$.

## Salidas Esperadas

- `P(exactamente 2 caras) = 0.3760`  
  (valor exacto: $\frac{3}{8} = 0.375$)

- `E[X] = 1.5002`  
  (valor exacto: $\frac{3}{2} = 1.5$)
