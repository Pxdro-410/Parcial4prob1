## Descripción del Problema

**Experimento base:** Una caja contiene 5 canicas rojas, 3 azules y 2 verdes. Se extraen dos canicas al azar **sin reemplazo**.

**Parte 1:**  
Estimar la probabilidad de que ambas canicas extraídas sean rojas.

---

**Parte 2 — Dos cajas:**  
Supongamos que hay dos cajas:
- **Caja 1:** 5 canicas rojas, 3 azules y 2 verdes.
- **Caja 2:** 2 canicas rojas, 5 azules y 3 verdes.

Se elige una caja al azar y se extraen dos canicas sin reemplazo, que resultan ser **una roja y una verde**.  
Estimar la probabilidad de que provengan de la Caja 1.

## Salidas Esperadas

**Parte 1:**
- `P(ambas rojas) = 0.2249`  
  (valor exacto: $\frac{2}{9} \approx 0.2222$)

**Parte 2:**
- `P(Caja 1 | una roja y una verde) = 0.6115`  
  (valor exacto: $\frac{5}{8} = 0.625$)

## Requisitos de Implementación

1. Fijar semilla aleatoria con valor **2026** antes de cualquier generación de números aleatorios.
2. Realizar **10,000 repeticiones** del experimento.
3. Imprimir respuestas como número decimal con **cuatro decimales**.

## Solución

```python
# Inciso a
# se realizan los imports necesarios
import random

# definimos la semilla y el numero de repeticiones para que los resultados sean reproducibles
random.seed(2026)
N_REPETICIONES = 10000

# se crea una lista de las canicas
# R = Roja, A = Azul, V = Verde
caja = ['Roja', 'Roja', 'Roja', 'Roja', 'Roja', 'Azul', 'Azul', 'Azul', 'Verde', 'Verde']

# se inicializan los contadores de coincidencias
coincidencias_2_rojas = 0

for i in range(N_REPETICIONES):
    # se eligen 2 canicas al azar de la caja SIN REEMPLAZO con sample
    canica1, canica2 = random.sample(caja, 2)
    
    # se cuenta cuantas veces se obtuvieron exactamente 2 canicas rojas y se añade al contador de coincidencias
    if canica1 == 'Roja' and canica2 == 'Roja':
        coincidencias_2_rojas += 1
        
Prob_2_rojas = coincidencias_2_rojas / N_REPETICIONES
print("Inciso a:")
print(f'P(2 rojas) = {Prob_2_rojas:.4f}')


# ---------------------------------------------------------------------------------------------------------------


N_REPETICIONES = 10000

# se crea una lista de las canicas
# R = Roja, A = Azul, V = Verde
caja1 = ['Roja', 'Roja', 'Roja', 'Roja', 'Roja', 'Azul', 'Azul', 'Azul', 'Verde', 'Verde']
caja2 = ['Roja', 'Roja', 'Azul', 'Azul', 'Azul', 'Azul', 'Azul', 'Verde', 'Verde', 'Verde']

# Contadores para la probabilidad condicional
una_roja_y_verde = 0  # Cuenta cuántas veces logramos sacar 1 Roja y 1 Verde
casos_caja1 = 0   # De esas veces, cuantas vinieron de la caja 1

for i in range(N_REPETICIONES):
    # se escoje una caja al azar ya sea la 1 o 2
    caja_elegida = random.choice([1, 2])
    
    # se eligen 2 canicas al azar de la caja elegida SIN REEMPLAZO con sample
    if caja_elegida == 1:
        extraccion = random.sample(caja1, 2)
    else:
        extraccion = random.sample(caja2, 2)
        
    # la condicion es sacar exactamente 1 Roja y 1 Verde
    extraccion_roja = extraccion.count('Roja')
    extraccion_verde = extraccion.count('Verde')
    
    if extraccion_roja == 1 and extraccion_verde == 1:
        una_roja_y_verde += 1
        
        # si se cumple la condicion, se valida que fue de la caja1 
        if caja_elegida == 1:
            casos_caja1 += 1

# se calcula la prob condicional con bayes: 
# P(Caja 1 | una roja y una verde) = (Casos favorables de Caja 1) / (Total de veces que salió roja y verde)
prob = casos_caja1 / una_roja_y_verde
print(" ")
print("Inciso b:")
print(f"P(Caja 1 | una roja y una verde) = {prob:.4f}")
```
