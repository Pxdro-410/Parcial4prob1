## Descripción del Problema

**Experimento base:** De una baraja estándar de 52 cartas se extraen dos cartas al azar **sin reemplazo**.

**Preguntas:**

**a.** Estimar la probabilidad de que ambas cartas sean ases.

**b.** Sean:
- $A$: la primera carta es un As.
- $B$: la segunda carta es un As.

Determinar si los eventos $A$ y $B$ son **independientes**.

## Salidas Esperadas

- `P(ambas ases) = 0.0036`  
  (valor exacto: $\frac{1}{221} \approx 0.004524$)

- `P(A) * P(B) = 0.0055`  
  (valor exacto: $\frac{1}{169} \approx 0.005917$)

- `Los eventos son independientes: False`

## Requisitos de Implementación

1. Fijar semilla aleatoria con valor **2026** antes de cualquier generación de números aleatorios.
2. Realizar **10,000 repeticiones** del experimento.
3. Imprimir respuestas como número decimal con **cuatro decimales**.

## Solución

importacion inicial 

```python
import random
import numpy as np

N_SIMULACIONES = 10000

# baraja
baraja = ([1] * 4) + ([0] * 48) # 4 ases y 48 cartas no ases

baraja_arr = np.array(baraja)
```

Ahora hago la extraccion de las cartas en 2 listas

```python
random.seed(2026)

lista_cartas1 = []
lista_cartas2 = []

for _ in range(N_SIMULACIONES):
    carta1, carta2 = random.sample(baraja, 2) # sample saca sin reemplazo
    
    lista_cartas1.append(carta1)
    lista_cartas2.append(carta2)
    

cartas1 = np.array(lista_cartas1)
cartas2 = np.array(lista_cartas2)

```

# INCISO A 

Estimar la probabilidad de que ambas cartas sean ases.


```python
incisoA = np.mean ((cartas1 == 1) & (cartas2 == 1))

print(f"P(carta1 = As y carta2 = As) = {incisoA:.4f}")
```

# INCISO B 

**b.** Sean:
- $A$: la primera carta es un As.
- $B$: la segunda carta es un As.

Determinar si los eventos $A$ y $B$ son **independientes**.


Por teoría, sabemos que 2 eventos son independientes si P(A | B) = P(A), o bien si P(A) * P(B) = P(A & B)

```python

interseccion = incisoA

multiplicacion = np.mean(cartas1 == 1) * np.mean(cartas2 == 1)

print(f"P(carta1 = As y carta2 = As) = {interseccion:.4f}")
print(f"P(carta1 = As) * P(carta2 = As) = {multiplicacion:.4f}")

son_independientes = round(interseccion, 4) == round(multiplicacion, 4) # comparo con 4 decimales redondeando

print(f"Los eventos son independientes: {son_independientes}")
```

### Debido a que el resultado de la intersección comparado con la multiplicación de la probabilidad de los eventos aislados no son iguales, podemos concluir que no son eventos independientes.
